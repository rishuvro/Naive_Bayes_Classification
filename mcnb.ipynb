```python
import numpy as np
import pandas as pd
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
corpus = ['I loved Crow', 'I hated Crow', 'Crow is a good bird', 'Its black and Dirty', 'Black is a Good Color.']
```

```python
vectorizer = CountVectorizer(analyzer = "word", 
                             lowercase=True, 
                             tokenizer = None, 
                             preprocessor = None, 
                             stop_words = None, 
                             max_features = 5000,
                             token_pattern='[a-zA-Z0-9$&+,:;=?@#|<>^*()%!-]+')
```

```python
wm = vectorizer.fit_transform(corpus)
print(wm.todense())
```

```python
print(wm.shape)
```

```python
vocabulary = vectorizer.vocabulary_
print(vocabulary)
```

```python
tokens = vectorizer.get_feature_names_out()
print(tokens)
```

```python
doc_names = ['Doc{:d}'.format(idx) for idx, _ in enumerate(wm)]
df = pd.DataFrame(data=wm.toarray(), index=doc_names,
                  columns=tokens)
```
